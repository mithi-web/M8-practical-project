```python
import cv2
import numpy as np

# Load image in grayscale
img = cv2.imread("C:\\qqqq\\picturesque-mountain-valley-on-sunny-260nw-2477194071.webp", cv2.IMREAD_GRAYSCALE)

# Apply binary threshold
_, bw = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

cv2.imshow("Black & White", bw)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Gaussian Blur
blurred = cv2.GaussianBlur(img, (5, 5), 0)

cv2.imshow("Blurred Image", blurred)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Canny Edge Detection
edges = cv2.Canny(img, 100, 200)

cv2.imshow("Edges", edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Using ORB (Oriented FAST and Rotated BRIEF)
orb = cv2.ORB_create()
keypoints, descriptors = orb.detectAndCompute(img, None)

# Draw keypoints
feature_img = cv2.drawKeypoints(img, keypoints, None, color=(225,0,0))

cv2.imshow("Features", feature_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Load two overlapping images
img1 = cv2.imread(r"C:\\qqqq\\picturesque-mountain-valley-on-sunny-260nw-2477194071.webp\\left.jpg")
img2 = cv2.imread("C:\\qqqq\\picturesque-mountain-valley-on-sunny-260nw-2477194071.webp")
```

```python
if img1 is None or img2 is None:
    print("Error: One or both images could not be loaded. Check file paths.")
else:
    # Optional: resize to same dimensions for consistency
    img1 = cv2.resize(img1, (800, 600))
    img2 = cv2.resize(img2, (800, 600))

    # Stitcher API
    stitcher = cv2.Stitcher_create()
    (status, panorama) = stitcher.stitch([img1, img2])

    if status == cv2.STITCHER_OK:
        cv2.imshow("Panorama", panorama)
        cv2.imwrite("panorama_output.jpg", panorama)  # save result
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    else:
        print("Error during stitching. Status code:", status)
```
